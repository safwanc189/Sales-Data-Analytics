```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(df['Purchase'])
plt.title('Boxplot of Purchase Amounts')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Purchase'], kde=True, bins=30)
plt.title('Distribution of Purchase Amounts')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
palette = sns.color_palette("tab20", 20)
sns.barplot(x='Occupation', y='Purchase', data=df,palette=palette)
plt.title('Average Purchase Amount by Occupation')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
city_counts = df['City_Category'].value_counts()

plt.pie(x=city_counts, labels=city_counts.index, autopct='%1.1f%%')
plt.title('Distribution of City Categories')
plt.show()
```

```python
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
age_order = {
    '0-17': 0,
    '18-25': 1,
    '26-35': 2,
    '36-45': 3,
    '46-50': 4,
    '51-55': 5,
    '55+': 6
}
df['Age'] = pd.Categorical(df['Age'], categories=age_order.keys(), ordered=True)

sns.lineplot(x='Age', y='Purchase', hue='Gender', data=df,palette={'M': 'Brown', 'F': 'purple'})
plt.title('Gender-wise Purchase Comparison')
plt.xlabel('Age')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
numeric_columns = ['Product_Category_1', 'Product_Category_2', 'Purchase']
corr_matrix = df[numeric_columns].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu')

plt.title('Correlation Heatmap of Product Categories and Purchase')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_sample = df.sample(n=10000, random_state=42)
selected_columns = ['Product_Category_1', 'Product_Category_2', 'Purchase']

sns.pairplot(df_sample[selected_columns + ['Gender']], hue='Gender', palette='Set2', diag_kind='kde')
plt.figure(figsize=(10,6))
plt.show()
```

```python
sns.lineplot(x='Age', y='Purchase', hue='Marital_Status', data=df,palette={0: 'Red', 1: 'Green'})
plt.title('Married or Non Married Purchase Comparison')
plt.xlabel('Age')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
df_sample = df.sample(n=10000, random_state=42)
selected_columns = ['Product_Category_1', 'Product_Category_2', 'Purchase']

sns.pairplot(df_sample[selected_columns + ['Age']], hue='Age', palette='Set2', diag_kind='kde')
plt.figure(figsize=(10,6))
plt.show()
```

```python
order = ['0', '1', '2', '3', '4+']

sns.boxplot(x='Stay_In_Current_City_Years', y='Purchase', data=df, palette='Set2', order=order)

plt.title('Distribution of Purchase Amount by Stay Duration')
plt.xlabel('Years in Current City')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='City_Category', y='Purchase', data=df)
plt.title('Purchase Amount by City category')
plt.xlabel('Years in Current City')
plt.ylabel('Purchase Amount')
plt.show()
```

```python

sns.lineplot(data=df, x='Product_Category_1', y='Purchase', label='Product_Category_1')

# Plotting the second line with label
sns.lineplot(data=df, x='Product_Category_2', y='Purchase', label='Product_Category_2')

plt.suptitle('Purchase Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Purchase Amount')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Stay_In_Current_City_Years', palette='Set2',order=order)

plt.title('Count of People by Stay Duration in Current City', fontsize=16)
plt.xlabel('Years in Current City', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```
